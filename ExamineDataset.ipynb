```python
# get values
values = df.values
X, y = values[:, :-1], values[:, -1]
```

```python
# gather details
n_rows = X.shape[0]
n_cols = X.shape[1]
classes = np.unique(y)
n_classes = len(classes)
```

```python
# summarize
print('N Examples: {}'.format( n_rows))
print('N Inputs: {}'.format( n_cols))
print('N Classes: {}'.format( n_classes))
print('Classes: {}'.format(list(classes)))
```

```python
# class breakdown
print('Class Breakdown:')
breakdown = ''
for c in classes:
    total = len(y[y == c])
    ratio = (total / float(len(y))) * 100
    print(' - Class %s: %d (%.5f%%)' % (str(c), total, ratio))
```
